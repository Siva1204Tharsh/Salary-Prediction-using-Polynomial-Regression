# Salary Prediction using Polynomial Regression

### 01.Problem Statement
 Predicting the salary of an employee based on their level on previous job

### 02.Collecting Data
 X-Level | Y-Salary

### 03.Load and summarize the data

```python
import pandas as pd
dataset=pd.read_csv("dataset.csv")
dataset.head()
dataset.describe()
```

### 04.Segregating the data into X and y

```python
X=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
print(X)
print(y)
```

```python
dataset.isnull().any()
```

### 05.Spilitting the data to train and test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

### 06.Finding linear or non-linear relationship between two variables

```python
from  sklearn.linear_model import LinearRegression
modellr = LinearRegression()
modellr.fit(X,y)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, y , color='red')
plt.plot(X, modellr.predict(X), color='blue')
plt.title('Salary Prediction using Polynomial Regression')  
plt.ylabel('Salary')
plt.xlabel('Level')

plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
modelpr=PolynomialFeatures(degree=3)
Xploy=modelpr.fit_transform(X)
```

```python
model=LinearRegression()
model.fit(Xploy,y)
```

```python
plt.scatter(X,y, color='red')
plt.plot(X,model.predict(Xploy), color='blue')
plt.ylabel('Salary')
plt.xlabel('Level')
plt.show()
```
